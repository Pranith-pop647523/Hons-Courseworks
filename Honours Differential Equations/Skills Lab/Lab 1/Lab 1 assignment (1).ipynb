## Task 1

Use `SymPy` to solve the differential equation $y' = (y-y^2)\cos(x)$, with $y(0)=0.5$, and plot the solution.

```python
import sympy as sym
sym.init_printing()
from IPython.display import display_latex


x = sym.symbols('x')
y = sym.Function('y')
eq1 = sym.Eq(y(x).diff(x),(y(x)-y(x)**2)*sym.cos(x))
print("The equation")
display_latex(eq1)


eq1sol = sym.dsolve(eq1, y(x),ics={y(0):0.5})

print("has solutions")
display_latex(eq1sol)
print("or equivalently")
display_latex(sym.simplify(eq1sol))

import sympy.plotting as sym_plot

sym_plot.plot(eq1sol.rhs, (x,-10,10),ylim=[0,1], xlabel = 'x', xlabel_loc = 'right', ylabel = 'y')
```

## Task 2

Use `SciPy`'s `odeint` function to solve the system of equations

\begin{align*}
\frac{dx}{dt} &= 0.4x - 0.8xy \\ \frac{dy}{dt}&=xy -1.2y
\end{align*}

Produce a plot of the solutions for $0\leq t\leq 15$ with initial conditions $x(0)=0.5$ and $y(0)\in\{0.25, 0.75, 1.25, \ldots, 3.25\}$.

How many curves do you expect to see plotted?

```python
import numpy as np
from scipy.integrate import odeint
from matplotlib import pyplot as plt
```

```python

def dy_dt(X, t):
    x,y=X
    eq1 = 0.4*x-0.8*x*y
    eq2 = x*y - 1.2*y
    return [eq1,eq2]


x0=0.50
y0=np.arange(0.25,3.75,.5)


for y in y0:
    initial_con = [x0,y]

    t = np.linspace(0, 15, 150)

    sol = odeint(dy_dt,initial_con, t)
    plt.plot(sol[:, 0], sol[:, 1], label=f'y0 = {y}')

plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

plt.show()



```

Expected to see 7 curves/rings
